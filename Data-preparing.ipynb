```python
import pandas as pd
import numpy as np
frame =  pd.DataFrame.from_csv('PE-TRG-Jan-Mar-2017.csv').reset_index()
```

```python
#get rows count by PLU
frame.groupby('PLU')[['PLU']].size().rename('counts').reset_index().sort_values(['counts'], ascending=[False])
```

```python
#get specific products
products = ['2548','2477','2480','2551','2637']
sales = frame[frame['PLU'].isin(products)]
sales
```

```python
sales.to_csv('sales-jan-mar-2017.csv')
```

```python
#get products info
sales.drop_duplicates(['RecipeName','PLU']).reset_index()[['RecipeName','PLU']].to_csv('products.csv')
```

```python
#get locations
sales.drop_duplicates(['Locationid']).reset_index()[['Locationid']].to_csv('locations.csv')
```
